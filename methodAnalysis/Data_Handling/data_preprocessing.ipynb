# Data Cleaning

### Evaluate Data Quality

```python
# Check for missing values (NaN)
print('Missing Values:\n', df[['User_Input', 'Label_1', 'Label_2']].isnull().sum())

# Check for empty strings
print('Empty Strings:\n', (df[['User_Input', 'Label_1', 'Label_2']] == '').sum())
```

```python
# Check length of user responses

df['User_Input_Length'] = df['User_Input'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
print("User_Input length stats:\n", df['User_Input_Length'].describe())
```

```python
# Check label_1 distribution

print('Label_1 counts:\n', df['Label_1'].value_counts())
```

```python
# Check label_2 distribution 

print('Label_2 counts:\n', df['Label_2'].value_counts())
```
